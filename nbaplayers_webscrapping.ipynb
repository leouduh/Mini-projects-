```python
from bs4 import BeautifulSoup              #for webscrapping
import requests                            #request uses REST APi
import pandas as pd                        #creating data frame for data analysis
```

```python
letters = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l",
         "m", "n", "o", "p", "q", "r", "s", "t", "u", "v", "w", "y", "z"] #there is an all aba and nba players page where players names are alphabetically
url = "https://www.basketball-reference.com/players/"                     #arranged. every nba player with last name starting from a is is the page 
r = requests.get(url)                                                     #https://www.basketball-reference.com/players/a/ similarly for b just replace a after player/a with b and so on...
active_players = []
points_per_game = []
players_links = []
```

```python
for letter in letters:                                      #for loop used to get link to everyplayers profile. on the website activate players  name
    r = requests.get(url + letter + "/")                    #are written in bold 
    soup1 = BeautifulSoup(r.content, "html.parser")
    soup1 = BeautifulSoup(soup1.prettify(), "html.parser")
    parent_y = soup1.findAll("strong")                      #returns a list of all hmtl bold tags under every alphabetical group a-z
    for element in parent_y:                                #iterate through the loop 
        x = element.find("a")                               #find the a tag and get the players link....link to all their profiles
        if x == None:
            pass                                            #the code in line 5 of this cell returns a list and some elemets are "None" because they are not bold
        else:                                               #if else this with that
            players_links.append(str(x["href"]))            #links of active nba players are only appended
```

```python
len(players_links)                                   #883 active players on the site 
```

```python
for link in players_links:                                                  #for loop to iterate over every active player
    r = requests.get("https://www.basketball-reference.com" + link)         #sends a request returns a response object
    soup2 = BeautifulSoup(r.content, "html.parser")                         #parse the html content of the repsonse
    player_page_header = soup2.find("h1")                                   #search for header h1 tag 
    player_name = player_page_header.get_text().strip()                     #get the text "the players name" of the header tag h1... use strip method to make it neat
    active_players.append(str(player_name))                                 #append players name 
    level2 = soup2.find("tfoot")                                            #find tfoot tag
    player_ppg = level2.findChild("td", {"data-stat": "pts_per_g"}).get_text().strip()    #get their career ppg stat
    points_per_game.append(float(player_ppg))                                              #append thier ppg to the list
```

```python
data_set = {"Player" : active_players, "Career_PPG": points_per_game}                   #create a dictionary 
data_set = pd.DataFrame(data_set)                                                       #create DataFrame
data_set.head(5)
```
